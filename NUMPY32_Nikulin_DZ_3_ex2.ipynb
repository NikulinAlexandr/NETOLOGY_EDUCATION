```python
import pandas as pd
geo_data = {
'Центр': ['москва', 'тула', 'ярославль'],
'Северо-Запад': ['петербург', 'псков', 'мурманск'],
'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}

def query_region(query_text):
    for reg in geo_data.keys():
        for a in geo_data[reg]:
            if a in query_text:
                return reg
    else: return 'undefined'

keywords = pd.read_csv('C:\\ml-latest-small\\keywords.csv')
keywords['region'] = keywords['keyword'].apply(query_region)
#Для проверки
#keywords.query("keyword in ['авито москва', 'москва', 'тула', 'ярославль', 'петербург', 'псков', 'мурманск', 'владивосток', 'сахалин', 'хабаровск']")
```
